### Looping Through the Strings

```python
i=0
while i < len(fruit):
    letter = fruit[i]
    print(i, letter)
    i = i + 1
    
```

```python
### Another Way to Loop
for letter in fruit:
    print(letter)
```

```python
### Looping and counting
count = 0
for letter in fruit :
    if letter == 'p' :
        count += 1
print(count)
```

### Strings are immutable that is they cannot be changed
#### replace will give you new string but won't make changes in the same string

```python
greet="Hello Bob"
meet = greet
greet = greet.replace("Bob", "John")
print(greet)
print(meet)
print(greet.replace("Bob", "John"))
```

```python
print(greet.replace('o', 'X'))
print(greet)
```

### Stripping WhiteSpaces

```python
greet='    Hello Bob    '
print(greet.lstrip())
print(greet.rstrip())
print(greet.strip())
```

### Parsing and Extracting

```python
data = 'From  naveen.aggarwal@pu.ac.in  Sat Jan 5 09:14:16 2018'
atpos = data.find('@')
print(atpos)
```

```python
sppos = data.find(' ',atpos)
print(sppos)
```

```python
host = data[atpos+1 : sppos]
print(host)
```
