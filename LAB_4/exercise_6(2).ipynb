### Q4. Find the rank of Panjab University as per rpc attribute.

```python
df_sorted = df.sort_values(by='rpc', ascending=False).reset_index(drop=True)
# df_sorted["rank"] = df
df_sorted["Rank"] = df_sorted.index + 1

college_rank = df_sorted[df_sorted["name"]== "Panjab University"]["rank"]
print(college_rank)
```

```python
rank = df.sort_values(by="rpc", ascending=False)\
         .reset_index(drop=True)\
         .index[df["name"] == "Panjab University"][0] + 1

print(rank)
```

### Q5. List the colleges whose rpc value is less than Panjab University and rank is higher(rank value is less).

```python
df1 = df.sort_values(by = "rpc",ascending=False ).reset_index(drop=True)

df1["rank"] = df1.index + 1
# print(df1)
pu = df1[df1["name"] == "Panjab University"]
pu_rpc = pu["rpc"].values[0]
pu_rank = pu["rank"].values[0]
print(pu_rpc)
print(pu_rank)

result = df1[
    (df1["rank"] < pu_rank) & (df1["rpc"] < pu_rpc)
]
print(result)
```

```python

# Sort and rank
df_sorted = df.sort_values(by="rpc", ascending=False).reset_index(drop=True)
df_sorted["rank"] = df_sorted.index + 1

# Get Panjab University data
pu = df_sorted[df_sorted["name"] == "Panjab University"]

pu_rpc = pu["rpc"].values[0]
pu_rank = pu["rank"].values[0]

# Filter
result = df_sorted[
    (df_sorted["rpc"] < pu_rpc) &
    (df_sorted["rank"] < pu_rank)
]

print(result)
```
