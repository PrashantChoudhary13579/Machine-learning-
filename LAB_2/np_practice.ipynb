```python
import numpy as np
arr = np.array([1,2,3,4,5])
print(arr)
```

```python
print(np.__version__)
print(type(arr))
print(arr.dtype)
```

```python
a = np.array((1,2,3,4,5,6,7))
print(a)
```

```python
# 0-D Array
a = np.array(21)
print(a)
print(type(a))
print(a.dtype)

# 1-D Array
b = np.array([1,2,3,4,5])
print(b)
print(type(b))
print(b.dtype)

# 2-D Array
c = np.array([[1,2,3],[4,5,6]])
print(c)
print(type(c))
print(c.dtype)

# 3-D Array
d = np.array([[[1,3],[3,4]],[[1,2],[9,1]]])
print(d)
```

```python
print(a.ndim)
print(b.ndim)
print(c.ndim)
print(d.ndim)
```
